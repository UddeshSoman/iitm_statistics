```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
df = data.sort_values('date')
df.head()
```

```python
df.drop('Unnamed: 0',axis = 1 , inplace = True)
df.head()
```

```python
plt.figure(figsize = (15,15))
sns.lineplot(df.date, df.death, hue = df.state)
plt.legend(loc= 'upper left')
plt.xscale("linear")
```

```python
plt.figure(figsize = (15,15))
sns.scatterplot(df.positive, df.hospitalized)
plt.xscale("linear")
```

```python
plt.figure(figsize = (15,15))
df = df.groupby(['state']).agg(negative=("negative", 'sum'))
df = df.reset_index()
sns.barplot(df.state, df.negative)
```
